# Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

### Identification of Potential Outliers:

From the boxplot, we can observe that there is one potential outlier, which is the Jet Airways Business ticket price. This price is significantly higher than the prices of all other airlines and falls outside of the whiskers in the boxplot.

Impact on Analysis:

Potential outliers can impact your analysis in the following ways:

Skewing Statistics: Outliers can skew summary statistics, such as the mean and standard deviation, making them less representative of the majority of the data. In this case, the Jet Airways Business ticket price can significantly inflate the average flight price.

Misleading Visualizations: Outliers can distort data visualizations. They might make it challenging to visualize the central tendency or trends in the data, especially in cases like boxplots where whiskers and quartiles are influenced.

Bias in Models: If the analysis involves machine learning or statistical modeling, outliers can influence model performance by affecting parameter estimation and prediction accuracy.

### Handling Potential Outliers:

The appropriate approach for handling potential outliers depends on the context of your analysis and the nature of the data:

Investigate the Outlier: It's important to investigate the Jet Airways Business ticket price to understand why it is exceptionally high. Determine whether it's a genuine data point or if it could be due to data entry errors or other factors.

Use Robust Statistics: Consider using robust statistical measures, such as the median instead of the mean, to describe central tendency, as these are less influenced by outliers.

Outlier Treatment: Depending on the investigation results, you may decide to remove the outlier if it is deemed an error or an extreme rarity. However, be cautious about removing data points, as it can introduce bias.

Transformation: You can apply data transformations (e.g., log transformation) to mitigate the impact of extreme values.

The choice of how to handle outliers should be made carefully, keeping the analysis goals and data integrity in mind.

## Q13. How does the category of app affect its price? Create a bar chart to compare average prices by app cateogry.

```python
import pandas as pd
df=pd.read_csv('google_playstore_dataset.csv')
print('dataset contains ',df.shape[0],'rows and ',df.shape[1],'columns')

import pandas as pd
import matplotlib.pyplot as plt

# Convert the 'Price' column to a numerical format, removing the dollar sign and converting to float
df['Price'] = df['Price'].apply(lambda x: float(x.strip('$')) if '$' in x else 0)

# Group the data by 'Category' and calculate the average price for each category
average_prices = df.groupby('Category')['Price'].mean()

# Create a bar chart
plt.figure(figsize=(12, 6))
average_prices.plot(kind='bar')
plt.title('Average Prices of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```

Observation: Categories like finance have high pricing, followed by lifestyle and medical categories, which also have high prices.

# Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
import pandas as pd
df=pd.read_csv('google_playstore_dataset.csv')
print('dataset contains ',df.shape[0],'rows and ',df.shape[1],'columns')

# Remove non-numeric characters from the 'Installs' column and convert it to integers
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '').str.replace('Free', '0').astype(int)

# Create a frequency table for apps based on the number of installs
app_installs_frequency = df.groupby('App')['Installs'].sum().reset_index()

# Sort the frequency table in descending order to find the top 10 most popular apps
top_10_apps = app_installs_frequency.sort_values(by='Installs', ascending=False).head(10)

# Display the top 10 most popular apps
print(top_10_apps)
```
